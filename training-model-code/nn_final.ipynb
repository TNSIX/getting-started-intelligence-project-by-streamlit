```python
# ติดตั้ง kagglehub
!pip install kagglehub
import kagglehub
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import os

# ดาวน์โหลด dataset
path = kagglehub.dataset_download("muratkokludataset/pistachio-image-dataset")
print("Dataset downloaded to:", path)

# ตรวจสอบโครงสร้างโฟลเดอร์
print("Files in dataset:", os.listdir(path))

# กำหนด path ไปยังโฟลเดอร์รูปภาพ
dataset_path = os.path.join(path, "Pistachio_Image_Dataset", "Pistachio_Image_Dataset")
print("Image folders:", os.listdir(dataset_path))

# ตั้งค่าพารามิเตอร์
IMG_HEIGHT = 128  # ปรับขนาดภาพ
IMG_WIDTH = 128
BATCH_SIZE = 16
NUM_CLASSES = 2  # 2 คลาส: Kirmizi_Pistachio และ Siirt_Pistachio

# สร้าง ImageDataGenerator
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# โหลดข้อมูลฝึก
train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

# โหลดข้อมูล validation
validation_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

# ตรวจสอบข้อมูล
print("Number of training samples:", train_generator.samples)
print("Number of validation samples:", validation_generator.samples)
print("Class indices:", train_generator.class_indices)

# สร้างโมเดล CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(NUM_CLASSES, activation='softmax')  # 2 คลาส
])
model.summary()

# Compile โมเดล
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# ฝึกโมเดล
history = model.fit(
    train_generator,
    steps_per_epoch=max(1, train_generator.samples // BATCH_SIZE),
    epochs=20,
    validation_data=validation_generator,
    validation_steps=max(1, validation_generator.samples // BATCH_SIZE)
)

# แสดงผลลัพธ์
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
import os
import matplotlib.pyplot as plt

# Function to load and preprocess a single image
def load_and_preprocess_image(img_path, target_size=(128, 128)):
    # Load the image
    img = image.load_img(img_path, target_size=target_size)
    # Convert to array
    img_array = image.img_to_array(img)
    # Rescale like in training (1./255)
    img_array = img_array / 255.0
    # Expand dimensions to match model input (add batch dimension)
    img_array = np.expand_dims(img_array, axis=0)
    return img_array

# Function to predict pistachio type
def predict_pistachio(model, img_path):
    # Preprocess the image
    processed_img = load_and_preprocess_image(img_path)

    # Make prediction
    prediction = model.predict(processed_img)

    # Get class names from train_generator
    class_names = list(train_generator.class_indices.keys())

    # Get predicted class and confidence
    predicted_class_idx = np.argmax(prediction[0])
    predicted_class = class_names[predicted_class_idx]
    confidence = prediction[0][predicted_class_idx] * 100

    return predicted_class, confidence

# Get the path to the dataset
dataset_path = os.path.join(path, "Pistachio_Image_Dataset", "Pistachio_Image_Dataset")

# Choose a folder (e.g., Kirmizi_Pistachio)
folder_path = os.path.join(dataset_path, "Kirmizi_Pistachio")

# Get list of image files
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Limit to 10 images
image_files = image_files[:10]

# Set up the plot
plt.figure(figsize=(15, 10))

# Loop through 10 images
for i, image_file in enumerate(image_files, 1):
    # Construct full path
    test_image_path = os.path.join(folder_path, image_file)

    # Make prediction
    predicted_class, confidence = predict_pistachio(model, test_image_path)

    # Print prediction
    print(f"Image {i}: {image_file}")
    print(f"Predicted class: {predicted_class}")
    print(f"Confidence: {confidence:.2f}%")
    print("-" * 30)

    # Display image in subplot
    plt.subplot(2, 5, i)  # 2 rows, 5 columns
    img = image.load_img(test_image_path)
    plt.imshow(img)
    plt.title(f"{predicted_class}\n({confidence:.2f}%)")
    plt.axis('off')

# Adjust layout and show all images
plt.tight_layout()
plt.show()
```
